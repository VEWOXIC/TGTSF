```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.DataFrame()
for yyy in range(2019, 2023):
    a = pd.read_csv(f'./weather_csv/mpi_roof_{yyy}a.csv')
    b = pd.read_csv(f'./weather_csv/mpi_roof_{yyy}b.csv')

    c = pd.concat([a, b], axis=0)

    all_data = pd.concat([all_data, c], axis=0)
```
