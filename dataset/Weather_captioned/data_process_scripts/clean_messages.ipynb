```python
import hashlib
# create file name
def get_file_name(sentence):
    h = hashlib.blake2b(digest_size=5)
    h.update(sentence.encode('utf-8'))
    return h.hexdigest()
```

```python
get_file_name('It\'s the early morning of a day in January. ')
```

```python
wm_all1={}
for i in range(len(wm_all)):
    wm_all1[str(wm_all['time'][i])]=[get_file_name(j.replace(' ','')) for j in json.loads(wm_all['v3'][i])]
```
