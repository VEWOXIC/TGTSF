```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = pd.read_csv('electricity.csv')
```

```python
b
```

```python
# substract the year in date by 4
b['date'] = pd.to_datetime(b['date'])
b['date'] = b['date'].apply(lambda x: x.replace(year=x.year-4))

```

```python
b
```

```python
# add column dayofweek according to the date as text
b['dayofweek'] = b['date'].dt.day_name()
```

```python
b
```

```python
b['holiday']=0
for i in range(1096):
    day_data = b['319'][24*i:24*(i+1)]
    if (day_data>600).sum():
        pass
    else:
        if b['dayofweek'][24*i] == 'Saturday' or b['dayofweek'][24*i] == 'Sunday':
            pass
        else: 
            b['holiday'][24*i:24*(i+1)] = 1
```

```python
b
```

```python
# find where the holiday is 1
b[b['holiday']==1]
```

```python
# save the data to a new csv
b.to_csv('electricity_days.csv', index=False)
```

```python
BERT_model= 'paraphrase-MiniLM-L6-v2' #'all-mpnet-base-v2'
import os
save_dir='./Elec_news-embedding-'+BERT_model
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
from sentence_transformers import SentenceTransformer
model = SentenceTransformer(BERT_model).to('cuda:0')
from tqdm import tqdm
```

```python

news_embedding = model.encode('Public Holiday')
np.save(os.path.join(save_dir,'News-Holiday.npy'), news_embedding)
news_embedding = model.encode('Monday')
np.save(os.path.join(save_dir,'News-Monday.npy'), news_embedding)
news_embedding = model.encode('Tuesday')
np.save(os.path.join(save_dir,'News-Tuesday.npy'), news_embedding)
news_embedding = model.encode('Wednesday')
np.save(os.path.join(save_dir,'News-Wednesday.npy'), news_embedding)
news_embedding = model.encode('Thursday')
np.save(os.path.join(save_dir,'News-Thursday.npy'), news_embedding)
news_embedding = model.encode('Friday')
np.save(os.path.join(save_dir,'News-Friday.npy'), news_embedding)
news_embedding = model.encode('Saturday')
np.save(os.path.join(save_dir,'News-Saturday.npy'), news_embedding)
news_embedding = model.encode('Sunday')
np.save(os.path.join(save_dir,'News-Sunday.npy'), news_embedding)
news_embedding = model.encode('Weekend')
np.save(os.path.join(save_dir,'News-Weekend.npy'), news_embedding)
news_embedding = model.encode('Weekday')
np.save(os.path.join(save_dir,'News-Weekday.npy'), news_embedding)
news_embedding = model.encode('Monday, Public Holiday')
np.save(os.path.join(save_dir,'News-Monday_Holiday.npy'), news_embedding)
news_embedding = model.encode('Tuesday, Public Holiday')
np.save(os.path.join(save_dir,'News-Tuesday_Holiday.npy'), news_embedding)
news_embedding = model.encode('Wednesday, Public Holiday')
np.save(os.path.join(save_dir,'News-Wednesday_Holiday.npy'), news_embedding)
news_embedding = model.encode('Thursday, Public Holiday')
np.save(os.path.join(save_dir,'News-Thursday_Holiday.npy'), news_embedding)
news_embedding = model.encode('Friday, Public Holiday')
np.save(os.path.join(save_dir,'News-Friday_Holiday.npy'), news_embedding)
```

```python
for i in range(321):
    desc_embedding = model.encode('Channel '+str(i))
    np.save(os.path.join(save_dir,'Channel-'+str(i)+'.npy'), desc_embedding)
```
